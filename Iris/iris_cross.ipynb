```python
import pandas as pd
from keras import optimizers
from keras.layers import Dense, Dropout
from keras.models import Sequential
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
import numpy as np
```

```python
repository = pd.read_csv("./Models/iris.csv")
```

```python
features = repository.iloc[:, 0:4].values
classes = repository.iloc[:, 4].values
labelEncoder = LabelEncoder()

classes = labelEncoder.fit_transform(classes)
classDummy = np_utils.to_categorical(classes)
# iris setosa     1 0 0
# iris virginica  0 1 0
# iris versicolor 0 0 1
```

```python
def createNetwork():
    classifier = Sequential()
    classifier.add(Dense(
        units=round((4+3)/2),
        activation='relu',
        input_dim = 4
    ))

    classifier.add(Dense(
        units=round((4+3)/2),
        activation='relu',
    ))

    classifier.add(Dense(
        units=3,
        activation='softmax',
    ))

    classifier.compile(optimizer='adam',
                       loss='categorical_crossentropy',
    )

    return classifier
```

```python
classifier = KerasClassifier(build_fn=createNetwork,
                             epochs=1000,
                             batch_size=10)

results = cross_val_score(estimator=classifier,
                          X=features, y=classes,
                          cv=10, scoring='accuracy')
```

```python
media=results.mean()
deviation=results.std()
```
